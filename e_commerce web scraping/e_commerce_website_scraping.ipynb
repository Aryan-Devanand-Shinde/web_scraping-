collecting data of devices and their info from an e_commerce website ("https://webscraper.io/test-sites")

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd 
import numpy as np
```

get the url and load the website and check the status code

```python
URL = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

web = rq.get(URL)
web.status_code
```

get the web content and create an object of bs

```python
web_code = web.content
```

```python
soup = bs(web.text,"html.parser")
soup
```

get info about name price info_details

```python
# get names using id or class
names_list = soup.find_all("a",class_="title")
```

```python
names_list_text = []
for title in names_list :
    names_list_text.append(title.text)
print(names_list_text)    
```

```python
price_list = soup.find_all("h4",class_="price float-end card-title pull-right")
```

```python
price_list_text=[]
for price in price_list:
    price_list_text.append(price.text)
print(price_list_text)    
```

```python
details_list = soup.find_all("p",class_="description card-text")
```

```python
details_list_text = []
for i in details_list:
    details_list_text.append(i.text)
print(details_list_text)    
```

```python
dict_for_tablets = { "names" : names_list_text , "price" : price_list_text , "description" : details_list_text }
```

```python
df = pd.DataFrame(dict_for_tablets)
```
